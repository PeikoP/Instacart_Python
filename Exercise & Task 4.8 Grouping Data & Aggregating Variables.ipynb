## Step 6 - Create a spending flag for each user based on the average price across all their orders

```python
# grouping average order prices by user_id and entering data into a newly created column 'average_spent'
ords_prods_merge['average_spent'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# creating spending flag
ords_prods_merge.loc[ords_prods_merge['average_spent'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spent'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.head()
```

```python
# checking work, spending flag counts
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

## Step 7 - Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# grouping median days_since_last_order by user_id and creating a column for the results
ords_prods_merge['median_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
# checking work
ords_prods_merge.head()
```

```python
# creating order_frequency_flag
ords_prods_merge.loc[ords_prods_merge['median_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_orders'] > 10) & (ords_prods_merge['median_orders'] <=20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# checking work
ords_prods_merge.head()
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
