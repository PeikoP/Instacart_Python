```python
# creating price_label() function with if/else statement
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range'
    elif row['prices'] > 15:
        return 'High-range'
    else: return 'Not enough data'
```

## Step 4 - Identify busiest hours of the day and create a new column labeled busiest_period_of_day that categorizes the periods of time by 'Most orders', 'Average orders', and 'Fewest orders'.

```python
# taking a look at the dataframe again
df_merged.head()
```

```python
# finding the frequency to create a baseline to measure all results by
df_merged['order_hour_of_day'].value_counts(dropna = False)
```

Observations: I took this frequency count and in excel divided the counts into quartiles to find logical ways of puting the data into three categories. There is a drop off in numbers after hour 19, which coincided with the median or Q2 in the data. I chose to put hours where the count was above the median into 'Most orders', and anything below Q1 in 'Fewest orders', with all other hours into the 'Average orders' category. 

```python
# creating new empty list for results
hour_result = []
```

```python
# creating for-loop to categorize order_hour_of_day column
for value in df_merged['order_hour_of_day']:
    if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9 or value == 17 or value == 8 or value == 18 or value == 19:
        hour_result.append('Most orders')
    elif value == 0 or value == 1 or value == 5 or value == 2 or value == 4 or value == 3:
        hour_result.append('Fewest orders')
    else:
        hour_result.append('Average orders')
```

```python
# creating new column and assigning hour_results to it
df_merged['busiest_period_of_day'] = hour_result
```
